```
# Define rooms, items, and monsters
# Doors
door_spooky = {
    "name": "door spooky room",
    "type": "door",
}
door_bloody = {
    "name": "door bloody room",
    "type": "door",
}
door_monster = {
    "name": "door monster room",
    "type": "door"
}


# Keys
key_monster = {
    "name": "monster key",
    "type": "key",
    "target": door_monster,
}
key_spooky = {
    "name": "spooky key",
    "type": "key",
    "target": door_spooky,
}
key_bloody = {
    "name": "bloody key",
    "type": "key",
    "target": door_bloody,
}

# Furnitures
candle = {
    "name": "candle",
    "type": "furniture",
}
mirror = {
    "name": "mirror",
    "type": "furniture",
}
bloody_handprint = {
    "name": "bloody handprint",
    "type": "furniture",
}
cobwebs = {
    "name": "cobwebs",
    "type": "furniture",
}
creepy_doll = {
    "name": "creepy doll",
    "type": "furniture",
}
torture_instruments = {
    "name": "torture instruments",
    "type": "furniture",
}
altar = {
    "name": "altar",
    "type": "furniture",
}
sarcophagus = {
    "name": "sarcophagus",
    "type": "furniture",
}
monster_claw_marks = {
    "name": "monster claw marks",
    "type": "furniture",
}
ominous_growling = {
    "name": "ominous growling",
    "type": "furniture",
}

# Rooms
confusion_room = {
    "name": "confusion room",
    "type": "room",
}
spooky_room_2 = {
    "name": "spooky room 2",
    "type": "room",
}
bloody_room = {
    "name": "bloody room",
    "type": "room",
}

bloody_room_2 = {
    "name": "bloody room 2",
    "type": "room",
}

monster_room = {
    "name": "monster room",
    "type": "room",
}
outside = {
    "name": "outside"
}
spooky_room = {
    "name": "spooky room",
    "type": "room",
}
game_over = {
    "name": "hell"
}

# Bones
bone_confusion = {
    "name": "bone from confusion room",
    "type": "bone",
    "target": "cerberus"
}
bone_spooky = {
    "name": "bone from spooky room",
    "type": "bone",
    "target": "cerberus"
}
bone_bloody = {
    "name": "bone from bloody room",
    "type": "bone",
    "target": "cerberus"
}

all_rooms = [confusion_room, spooky_room, bloody_room, monster_room, outside, spooky_room_2, bloody_room_2]

all_doors = [door_spooky, door_bloody, door_monster]

all_keys = [key_monster, key_spooky, key_bloody]

all_furnitures = [candle, mirror, bloody_handprint, cobwebs, creepy_doll, torture_instruments, altar, sarcophagus, monster_claw_marks, ominous_growling]

all_bones = [bone_confusion, bone_spooky, bone_bloody]

# Define relations between rooms and items
object_relations = {
    "confusion room": [candle, mirror, door_spooky],
    "spooky room 2": [bloody_handprint, cobwebs, creepy_doll, door_spooky, door_bloody],
    "bloody room 2": [torture_instruments, altar, door_bloody, door_monster],
    "monster room": [sarcophagus, monster_claw_marks,door_bloody, door_monster],
    "outside": [],
    "door spooky room": [confusion_room, spooky_room],
    "door bloody room": [bloody_room, spooky_room_2],
    "door monster room": [outside, monster_room, bloody_room],
    "door spooky room 2": [spooky_room, spooky_room_2, bloody_room],
    "door bloody room 2": [bloody_room, bloody_room_2, monster_room],
    "candle": [key_spooky],
    "mirror": [bone_confusion],
    "cobwebs": [bone_spooky],
    "creepy doll":[key_bloody],
    "altar": [key_monster],
    "torture instruments": [bone_bloody],
    "spooky room": [door_spooky, door_bloody],

}

# Define game state
INIT_GAME_STATE = {
    "current_room": confusion_room,
    "keys_collected": [],
    "bones_collected": [],
    "target_room": outside,
    "game_over": game_over,
    "monster_awake": False,
}

def linebreak():
    """
    Print a line break
    """
    print("\n\n")

def start_game():
    """
    Start the game
    """
    print("You wake up in a dark, damp room. Your head is spinning and you can't remember how you got here. You see a faint light coming from a doorway ahead.")
    play_room(game_state["current_room"])

def play_room(room):
    """
    Play a room. First check if the room being played is the target room.
    If it is, the game will end with success. Otherwise, let player either 
    explore (list all items in this room) or examine an item found here.
    """
    game_state["current_room"] = room
    if game_state["current_room"] == game_state["target_room"]:
        if len(game_state["bones_collected"]) == 3:
            print("You offer the bones to the Cerberus. It accepts your offering and lets you pass. Congratulations! You've escaped the haunted castle!")
        else:
            print("You step into the monster room, but the Cerberus growls menacingly. You haven't collected all the bones yet. It pounces on you, and you become its next meal. Game over!")
            game_state["current_room"] = game_over
            if game_state["current_room"] == game_state["game_over"]:
                print("LOSERRRRRR. Try again!")
                player_choice = input("Restart ? ('Y' or 'N'): ")
                if player_choice == 'Y':
                    start_game()
                else :
                    exit()
    else:
        print("You are now in " + room["name"])

        if room["name"] == "spooky room":
            spooky_room_game(room)

        if room["name"] == "bloody room":
            bloody_room_game(room)    

        intended_action = input("What would you like to do? Type 'explore' or 'examine': ").strip()
        if intended_action == "explore":
            explore_room(room)
            play_room(room)
        elif intended_action == "examine":
            examine_item(input("What would you like to examine? ").strip())
        else:
            print("Not sure what you mean. Type 'explore' or 'examine'.")
            play_room(room)
        linebreak()


def explore_room(room):
    """
    Explore a room. List all items belonging to this room.
    """
    items = [i["name"] for i in object_relations[room["name"]]]
    ambiance = {
        "confusion room": "The confusion room is in almost pitch black darkness. You can barely see anything.",
        "spooky room": "The spooky room is filled with eerie shadows and strange whispers echo off the walls.",
        "bloody room": "The bloody room has gruesome sights, with bloodstains splattered across the walls and floor.",
        "monster room": "The monster room is suffused with an oppressive atmosphere, and you can feel the presence of something malevolent lurking nearby."
    }
    print(ambiance.get(room["name"], "You explore the room.") + " You find " + ", ".join(items))

def get_next_room_of_door(door, current_room):
    """
    From object_relations, find the room connected to the given door.
    """
    connected_rooms = object_relations[door["name"]]
    for room in connected_rooms:
        if(not current_room == room):
            return room

def examine_item(item_name):
    """
    Examine an item which can be a door or furniture.
    First make sure the intended item belongs to the current room.
    Then check if the item is a door. Tell player if key hasn't been 
    collected yet. Otherwise ask player if they want to go to the next
    room. If the item is not a door, then check if it contains keys.
    Collect the key if found and update the game state. At the end,
    play either the current or the next room depending on the game state
    to keep playing.
    """
    current_room = game_state["current_room"]
    next_room = ""
    output = None
    
    for item in object_relations[current_room["name"]]:
        if(item["name"] == item_name):
            output = "You examine " + item_name + ". "
            if(item["type"] == "door"):
                have_key = False
                for key in game_state["keys_collected"]:
                    if(key["target"] == item):
                        have_key = True
                if(have_key):
                    output += "You unlock it with a key you have."
                    next_room = get_next_room_of_door(item, current_room)
                else:
                    output += "It is locked but you don't have the key."
            else:
                if(item["name"] in object_relations):
                    items_in_item = object_relations[item["name"]]
                    for element in items_in_item:
                        if element['type'] == 'key':
                            game_state["keys_collected"].append(element)
                            output += "You find " + element["name"] + "."
                        elif element['type'] == 'bone':
                            game_state["bones_collected"].append(element)
                            output += "You find " + element["name"] + "."
                    object_relations[item["name"]] = []
                else:
                    output += "There isn't anything interesting about it."
            print(output)
            break


    if output is None:
        print("The item you requested is not found in the current room.")
    
    if next_room and input("Do you want to go to the next room? Enter 'yes' or 'no': ").strip() == 'yes':
        play_room(next_room)
    else:
        play_room(current_room)


def spooky_room_game(current_room):
    """
    Play a special game in the spooky room.
    """
    print("You are in a room filled with fear. A ghost is sitting in the middle of the room.")
    print("He asks you a riddle:")
    print(""" 
    "Three heads I have, and each a different face,
    Guarding the gates to the darkest place.
    With a bone in hand, my loyalty you gain,
    But beware, for trespassed, you'll feel my bane.
    What am I?" """)
    
    while True:
        riddle_input = input("Answer the riddle: ").strip().lower()
        if riddle_input == "cerberus":
            print("You have answered correctly. The ghost disappears and you can stay in this room.")
            # Allow player to explore or examine the room
            play_room(get_next_room_of_door({"name": "door spooky room 2"}, current_room))
        else:
            print("Wrong Answer. Back to previous room.")
            play_room(get_next_room_of_door({"name": "door spooky room"}, current_room))
            break


def bloody_room_game(current_room):
    import random

    print("You have to win rock scissors paper against the vampire, if you want to survive and escape the hunted castle")
    print("Choose your move:")
    print("1. Rock")
    print("2. Paper")
    print("3. Scissors")
   
    moves = ["Rock", "Paper", "Scissors"]
    player_choice = int(input("Enter the number of your move: "))
   
    if player_choice not in [1, 2, 3]:
        print("Please enter a valid number.")
        bloody_room_game(current_room)
    else:
        player_move = moves[player_choice - 1]
        vampire_move = random.choice(moves)
       
        print("The vampire chooses:", vampire_move)

       
        if player_move == vampire_move:
            print("It's a tie!")
            bloody_room_game(current_room)
        elif (player_move == "Rock" and vampire_move == "Scissors") or \
             (player_move == "Paper" and vampire_move == "Rock") or \
             (player_move == "Scissors" and vampire_move == "Paper"):
            print("Congratulations, you have defeated the vampire! You can now move to the next room.")
            play_room(get_next_room_of_door({"name": "door bloody room 2"}, current_room))
        else:
            print("The vampire wins! You are now his blood slave for eternity.Game over!")
            game_state["current_room"] = game_over
            if game_state["current_room"] == game_state["game_over"]:
                print("LOSERRRRRR. Try again!")
                player_choice = input("Restart ? ('Y' or 'N'): ")
                if player_choice == 'Y':
                    start_game()
                else :
                    exit()
                    
```
